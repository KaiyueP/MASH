```python
#!/usr/bin/env python3
"""
Read and plot one-QD LVC excited-state populations from pop.out.

State ordering:
- States 1-9: QD exciton states
"""

import numpy as np
import matplotlib.pyplot as plt

autos = 2.4188843265857e-17
n_excited_states = 9

data = np.loadtxt("pop.out")
time = data[:, 0] * autos * 1e15  # au to fs
populations = data[:, 1:]
n_states = populations.shape[1]

if n_states != n_excited_states:
    raise ValueError("Expected %d QD excited states, found %d" % (n_excited_states, n_states))

fig, ax = plt.subplots(figsize=(8, 5))

for state in range(n_excited_states):
    ax.plot(time, populations[:, state], label="Exciton %d" % (state + 1), linewidth=1.5, alpha=0.8)

ax.set_xlabel("Time (fs)", fontsize=12)
ax.set_ylabel("Population", fontsize=12)
ax.set_title("One-QD exciton populations", fontsize=14, fontweight="bold")
ax.legend(ncol=3, fontsize=8)

plt.tight_layout()
plt.show()

print("Max excited-state populations:")
for state in range(n_excited_states):
    print("  Exciton %d: %.6g" % (state + 1, populations[:, state].max()))

```
